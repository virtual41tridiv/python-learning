## Tuples

```python
def find_pe_and_pb (price,eps,book_value):
    pe = price/eps
    pb = price/book_value
    return pe,pb
```

```python
pe_ratio, pb_ratio = find_pe_and_pb(150,5,25)
```

```python
pe_ratio
```

```python
pb_ratio
```
