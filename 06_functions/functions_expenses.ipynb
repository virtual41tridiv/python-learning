## FUNCTIONS

```python
expenses_sergey = [30,45,70,90]
expenses_sundar = [40,23,50,60]
```

```python
total_expenses_sergey = 0
total_expenses_sundar = 0
```

```python
for expense in expenses_sergey:
    total_expenses_sergey += expense

print("Total expenses for Sergey:", total_expenses_sergey)
```

```python
for expense in expenses_sundar:
    total_expenses_sundar += expense

print("Total expenses for Sundar:", total_expenses_sundar)
```

```python
def find_total(expenses):
    total = 0
    for expense in expenses:
        total += expense
    return total

total_expenses_sergey = find_total(expenses_sergey)
total_expenses_sundar = find_total(expenses_sundar)

print("Total expenses for Sergey:", total_expenses_sergey)
print("Total expenses for Sundar:", total_expenses_sundar)
```
