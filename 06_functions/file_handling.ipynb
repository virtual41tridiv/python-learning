## File Handling
In Python, you can handle files using built-in functions. Here are some common operations:

```python
f=open("funny.txt","r")

for line in f:
    print(line)
f.close()

```

```python
with open("funny.txt","r") as f:
    for line in f:
        print(line)
```

```python
with open("funny.txt","r") as f:
    lines = f.readlines()
    print(lines)
```

```python
with open("love.txt","w") as f:
    f.write("I love Python\n")
    f.write("I love programming\n")
    f.write("I love coding\n")
```

```python
with open("python.txt","w") as f:
    f.writelines([
        "Python is great\n",
        "Python is easy to learn\n",
        "Python is versatile\n"
        ])
```

```python
with open("scores.csv",'r') as f:
    for line in f:
        print(line)
```

```python
player_scores = {}
with open("scores.csv",'r') as f:
    for line in f:
        player,_,score = line.split(",")
        score = int(score.strip())
        

        if player in player_scores:
            player_scores[player].append(score)
        else:
            player_scores[player] = [score]
print(player_scores)
```

```python
for player,score_list in player_scores.items():
    min_score = min(score_list)
    max_score = max(score_list)
    average_score = sum(score_list)/len(score_list)
    print(f"{player}: {average_score}")
```
