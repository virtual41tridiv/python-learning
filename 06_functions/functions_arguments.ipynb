## Function Arguments

```python
def sum_all_num(*args):
    total=0
    for num in args:
        total += num
    return total

total = sum_all_num(1, 2, 3, 4, 5)
print(total)
```

```python
def company_info(**kwargs):
    for key in kwargs:
        print(key,kwargs[key])
company_info(name="Google", location="Mountain View", founded=1998,ticker="AAPL",CEO="Sundar Pichai",pe=30.5)
```
