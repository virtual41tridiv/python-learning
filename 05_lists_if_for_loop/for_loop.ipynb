## FOR LOOP

```python
expenses = [1200,1300,1500,1700]

total_expense = 0

for expense in expenses:
    total_expense = total_expense + expense

print("Total expense is: ", total_expense)
```

```python
expenses = [1200,1300,1500,1700]

total_expense = 0

for i in range(len(expenses)):

    expense = expenses[i]
    print(f"Month {i+1} expense is: {expense}")
    total_expense = total_expense + expenses[i]

print(f"Total expense is: {total_expense}")
```

```python
expenses = [1200,1300,1500,1700]

total_expense = 0

for i, expense in enumerate(expenses):
    print(f"Month {i+1} expense is: {expense}")
    total_expense = total_expense + expense

print(f"Total expense is: {total_expense}")
```

## Break in for loop

```python
monthly_sales = [42,38,33,38,40,45]
threshold = 35

for sales_amount in monthly_sales:
    if sales_amount < threshold:
        print(f"Sales amount {sales_amount} is below the threshold")
        break
    else:
        print(f"Sales amount {sales_amount} is above the threshold")
```

```python
monthly_sales = [42,38,33,38,40,45]
threshold = 35

for sales_amount in monthly_sales:
    if sales_amount < threshold:
        print(f"Sales amount {sales_amount} is below the threshold")
        #break
    else:
        print(f"Sales amount {sales_amount} is above the threshold")
```

```python
monthly_sales = [42,38,33,38,40,45]
month = ["January", "February", "March", "April", "May", "June"]
threshold = 35

for sales_amount, month in zip(monthly_sales, month):
    if sales_amount < threshold:
        print(f"Sales amount {sales_amount} in {month} is below the threshold")
        break
    else:
        print(f"Sales amount {sales_amount} in {month} is above the threshold")
```

```python
monthly_sales = [42,38,33,38,40,45]
month = ["January", "February", "March", "April", "May", "June"]
threshold = 35

for sales_amount, month in zip(monthly_sales, month):
    if sales_amount < threshold:
        print(f"Sales amount {sales_amount} in {month} is below the threshold")
        # break
    else:
        print(f"Sales amount {sales_amount} in {month} is above the threshold")
```
