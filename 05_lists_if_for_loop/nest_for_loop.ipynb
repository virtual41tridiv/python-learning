## Nested For Loop

```python
products = ["iphone","ipad","macbook","apple watch"]
regions = ["india","usa","uk","australia"]
revenue = [20,23,45,18,17,12,12,9,5,60,45,23,12,34,56,78]
```

```python
for product in products:
    for region in regions:
        print(f"{product} is sold in {region} ")
```

```python

i = 0
for product in products:
    for region in regions:
        rev = revenue[i]
        i = i + 1
        print(f"{product} {region} revenue is:", rev)
```
